```python
channel_conversion = (
    mc.groupby('CampaignChannel')['Conversion']
      .mean()
      .reset_index()
)

channel_conversion
```

```python
channel_conversion['ConversionRate'] = (
    channel_conversion['Conversion'] * 100
)

channel_conversion
```

```python
import warnings
warnings.filterwarnings('ignore')

colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.figure(figsize=(8,5))

ax = sns.barplot(
    data=channel_conversion,
    x='CampaignChannel',
    y='ConversionRate',
    palette=colors
)

# Add percentage labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%')

plt.title('Average Conversion Rate by Marketing Channel')
plt.xlabel('Marketing Channel')
plt.ylabel('Conversion Rate (%)')

plt.show()
```

```python
Website Funnel Analysis

Important variables:

WebsiteVisits
PagesPerVisit
TimeOnSite

Compare them against Conversion.
```

```python
metrics = [
    'WebsiteVisits',
    'PagesPerVisit',
    'TimeOnSite'
]

for col in metrics:
    
    ax = sns.boxplot(
        x='Conversion',
        y=col,
        data=mc,
        palette=['red', 'green']
    )
    
    # Calculate and display median values
    medians = mc.groupby('Conversion')[col].median()

    for i, median in enumerate(medians):
        ax.text(
            i,
            median,
            f'{median:.1f}',
            ha='center',
            va='bottom',
            fontweight='bold'
        )

    plt.title(f'{col} by Conversion Status')
    plt.show()
```

```python
Insight:

Customers spending more time on the website may convert more frequently.
```

```python
channel_summary = (
    mc.groupby('CampaignChannel')
      .agg(
          Total_Customers=('Conversion', 'count'),
          Total_Conversions=('Conversion', 'sum'),
          Avg_AdSpend=('AdSpend', 'mean')
      )
      .reset_index()
)

channel_summary['Conversion_Rate_%'] = (
    channel_summary['Total_Conversions']
    / channel_summary['Total_Customers']
) * 100

channel_summary.round(2)
```
